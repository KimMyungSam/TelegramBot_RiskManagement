요청 개수로 일봉 데이터 얻어오기
이번에는 일별 시가, 고가, 저가, 종가, 거래량과 같은 데이터를 API를 사용해서 얻는 방법을 살펴보도록 하겠습니다. 기술적 분석을 하는 경우 일봉 데이터가 잘 사용되며 특히 종가의 경우 수정 종가를 사용합니다. 또한, 기본적 분석을 하는 경우에도 과거 거래일 기준으로 시가총액이 필요한 경우가 많은데 API를 사용해서 같이 얻어와 보도록 하겠습니다.

크레온 API에서 종목별로 차트 데이터를 얻어올 때 사용하는 클래스는 StockChart 입니다. StockChart 클래스를 사용하여 일별 데이터 뿐만 아니라 주, 월, 분, 틱 (거래가 체결되는 시간)단위의 데이터도 요청할 수 있습니다. 그러나 분이나 틱 단위의 데이터는 과거 일부 기간 동안만 제공해줍니다. StockChart 클래스는 몇 가지 메서드를 제공하는 그 중 주요 메서드는 다음과 같습니다. 

1	SetInputValue	요청 데이터에 대한 정보를 설정	커피 주문 (아메리카노 1잔, 카페라때 2잔)
2	BlockRequest	데이터 요청	결재
3	GetHeaderValue	요청된 데이터에 대한 정보 확인	진동벨이 울리면 번호표 확인하고 주문 내역이 맞는지 확인
4	GetDataValue	요청된 데이터를 가져가기	주문한 내역이 맞으면 커피 가져가기


StockChart 클래스의 메서드를 설명하기 전에 먼저 스타벅스에서 커피를 주문하고 커피를 받는 과정을 살펴봅시다. 커피를 주문하려면 주문 데스크에 가서 아메리카노 1잔 그리고 카페라때 2잔과 같이 구체적이 주문 내역을 말해줘야 합니다. 그리고나서 결재를 하지요. 결재 후 직원으로부터 진동벨을 받게 되는데 진동벨이 울리면 우리는 주문 번호를 확인하고 주문 내역을 한 번 확인합니다. (아메리카노 1잔 그리고 카페라때 2잔이 맞는지) 주문 내역이 맞다면 주문한 커피를 가져가면 됩니다. 이때 쟁반이 없다면 아마도 한잔씩 커피를 가져가겠지요?

StockChart  클래스의 메서드도 위와 같은 순서로 호출해야 합니다. 먼저 SetInputValue 메서드를 호출하여 어떤 종목에 대한 데이터를 요청하는지 그리고 일봉 데이터 중 어떤 데이터를 요청하는지에 대한 값을 세팅해줘야 합니다. 이 과정은 커피숍에서 주문 내역을 적는 것과 비슷합니다. 주문 내역을 다 준비했다면 실제로 주문을 해야 하지요? 이때 BlockRequest 메서드를 호출합니다. BlockRequest 메서드를 호출한 후에는 GetHeaderValue를 호출하여 서버로부터 데이터가 몇 개가 있는지 확인해야 합니다. (커피가 몇 잔 나왔는지 확인하여 내가 주문한 것과 같은지를 확인하는 과정) 서버로부터 전송받은 데이터의 개수만큼 GetDataValue를 호출하여 데이터를 가져가면 됩니다. 

 


```python
import win32com.client
stock_chart = win32com.client.Dispatch("CpSysDib.StockChart")

stock_chart.SetInputValue(0, 'A233740')                     # 종목 코드
stock_chart.SetInputValue(1, ord('2'))                      # 요청 구분 '1': 기간, '2': 개수
stock_chart.SetInputValue(4, 100)                           # 요청 데이터의 개수
stock_chart.SetInputValue(5, [0, 2, 3, 4, 5, 8, 13])        # 요청 내용, 0-day, 2-open, 3,-high..
stock_chart.SetInputValue(6, ord('D'))                      # 일봉 데이터
stock_chart.SetInputValue(9, '1')                           # 수정 주가
stock_chart.BlockRequest()

count = stock_chart.GetHeaderValue(3)


for i in range(count):
    day = stock_chart.GetDataValue(0, i)  # 위 SetInputValue(5, [0, 2, 3, 4, 5, 8, 13]) 결과를 순서대로 get
    open = stock_chart.GetDataValue(1, i)
    high = stock_chart.GetDataValue(2, i)
    low = stock_chart.GetDataValue(3, i)
    close = stock_chart.GetDataValue(4, i)
    volume = stock_chart.GetDataValue(5, i)
    cap = stock_chart.GetDataValue(6, i)
    print(day, open, high, low, close, volume, cap)
```
